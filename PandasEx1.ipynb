```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df["job_posted_date"])
```

```python
us_jobs = df[df['job_country'] == "United States" ]

us_jobs = us_jobs[us_jobs["salary_year_avg"].notna()]

us_jobs.groupby("job_title_short")['salary_year_avg'].agg(["min", "max", "median", "count"]).sort_values(by='median')
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month #gets months from dates
monthly_counts = df.job_posted_month.value_counts() #gets the count of jobs in a month
monthly_counts = monthly_counts.sort_index() #sorts the count according to months
monthly_counts
```

```python
plt.plot(monthly_counts.index, monthly_counts)
```
