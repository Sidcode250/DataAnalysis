```python
def least_diff(a,b,c):
    diff1 = abs(a-b)
    diff2 = abs(b-c)
    diff3 = abs(c-a)

    return min(diff1,diff2,diff3)

print(least_diff(10,15,30))
```

```python
def mbf(x):
    return 5*x

def call(fn, arg):
    return fn(arg)

def mulcall(fn, arg):
    return fn(fn(arg))

print(call(mbf, 1),
mulcall(mbf, 1))
```

```python
def mod_5(x):
    return x%5

print(max(100,51,14,key=mod_5))
```
